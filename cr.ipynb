```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
# Load the dataset (for example, 'crop_recommendation.csv')
df = pd.read_csv('cr_dataset.csv')

# Display the first few rows of the dataset
print(df.head())
```

```python
# Check the shape of the dataset
print(df.shape)

# Check the data types and missing values
print(df.info())

# Summary statistics of the numerical columns
print(df.describe())
```

```python
# Extract features (X) and target (y)
X = df.drop('label', axis=1)  # Features (drop the target column, assuming it’s called 'label')
y = df['label']  # Target

# Standardize the features for better model performance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Check the shape of the split data
print(X_train.shape, X_test.shape)
```

```python
 #Initialize the RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model using the training data
rf_model.fit(X_train, y_train)
```

```python
# Predict on the test data
y_pred = rf_model.predict(X_test)
```

```python
# Sample input (N, P, K, temperature, humidity, pH, rainfall)
sample_input = np.array([[90,42,43, 20, 82, 5, 200]])

# Scale the input data
sample_input_scaled = scaler.transform(sample_input)

# Predict the crop
predicted_crop = rf_model.predict(sample_input_scaled)

# Display the predicted crop
print(f"The recommended crop is: {predicted_crop[0]}")
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Display a classification report for more detailed metrics
print(classification_report(y_test, y_pred))
```

```python
import joblib

# Save the model
joblib.dump(rf_model, 'crop_recommend_model.pkl')

# Save the scaler
joblib.dump(scaler, 'scal.pkl')

print(type(scaler))  
```
